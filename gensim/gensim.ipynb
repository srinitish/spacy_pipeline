```python
import spacy
nlp = spacy.load('en_core_web_sm')
def preprocess(text):
    doc = nlp(text)
    filtered_text =[]
    for text in doc:
        if(text.is_stop or text.is_punct):
            continue
        filtered_text.append(text.lemma)
    return wv.get_mean_vector(filtered_text)
```

```python
v =preprocess("Hi this is nitish")
v.shape
```

```python
df['vector_num'] = df['Text'].apply(lambda text: preprocess(text))
df.head()
```
