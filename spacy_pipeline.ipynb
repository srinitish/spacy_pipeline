```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm") 
doc = nlp("hello everyone this is spacy")
```

```python
nlp.pipe_names
```

```python
for token in doc:
    print(token, " | ", token.lemma_)
```

```python
doc = nlp("Anthropic has accidentally leaked the internal source code for its AI coding tool, Claude Code, exposing 2,000 lines of proprietary data and prompting user concerns regarding data security.")
for token in doc:
    print(token, " | ", token.lemma_)
```

```python
from spacy import displacy
displacy.render(doc,style = 'ent')
```

```python
import nltk
from nltk import PorterStemmer
doc = nlp(['hello','running','rounded','stressed','found']
stemmer = PorterStemmer()
for token in doc:
    print(token, " | ", stemmer.stem(token))
```

```python
doc = nlp('hello running rounded stressed found')
for token in doc:
    print(token, " | ", token.lemma_)
```
